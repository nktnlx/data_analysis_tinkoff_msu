#### Задание 2
Прометьте событиями экран сервисы (опишите события и их свойства)
Какая метрика/метрики удобства экрана?

#### Решение

##### Сервисы 1
1. Событие: Пользователь обратился к голосовому ассистенту. Метрика: Как часто пользуются голосовым ассистентом.
2. Событие: Пользователь обратился к поисковой строке. Метрика: Как часто пользуются поисковой строкой; сколько времени уходит на осуществление нужного поискового запроса; насколько точными (верными) оказываются всплывающие подсказки.  
3. Событие: Пользователь выбирает нужную ему иконку сервиса. Метрика: время на нахождение нужной иконки (правильно ли отранжированы иконки); сколько пролистываний ленты иконок совершает пользователь перед выбором нужного ему сервиса.  
4. Событие: Пользователь просматривает свои заказы. Метрика: Какую по счету плашку с заказом он выбирает (верно ли наше ранжирование заказов пользователя); как часто пользователь нажимаем на надпись "Все".  
5. Событие: Пользователь обратился к выбору города. Метрика: как часто пользователь изменяет город для которого хочет просмотреть набор сервисов. 
6. Событие: Пользователь обратился за технической поддержкой в чат. Метрика: как часто пользователь использует техническую поддержку.  
7. Событие: Пользователь провалился дальше по воронке и начал взаимодействие с экраном следующего уровня. Метрика: конверсия переходов дальше по воронке.

##### Сервисы 2
1. Событие: Пользователь обратился к выбору города. Метрика: как часто пользователь изменяет город для которого хочет просмотреть набор сервисов. 
2. Событие: Пользователь выбирает нужную ему иконку сервиса. Метрика: время на нахождение нужной иконки (правильно ли отранжированы иконки); сколько пролистываний ленты иконок совершает пользователь перед выбором нужного ему сервиса.  
3. Событие: Пользователь просматривает свои заказы. Метрика: Какую по счету плашку с заказом он выбирает (верно ли наше ранжирование заказов пользователя); как часто пользователь нажимаем на надпись "Все".  
4. Событие: Пользователь обратился к услуге "Подберем вам ресторан". Метрика: Как часто пользуются данной услугой; сколько времени затрачивается на подбор; конверсия в бронь столика. 
5. Событие: Пользователь начал листать карусель Лучших ресторанов. Метрика: конверсия в выбор ресторана, какой по счету ресторан в карусели кликнет пользователь, как часто пользователь нажимаем на надпись "Все" (удобнее просматривать списком, а не каруселью).  
6. Событие: Пользователь обратился за технической поддержкой в чат. Метрика: как часто пользователь использует техническую поддержку.  
7. Событие: Пользователь провалился дальше по воронке и начал взаимодействие с экраном следующего уровня. Метрика: конверсия переходов дальше по воронке.

#### Задание 3
Предложите минимум 3 варианта рекомендации действия клиенту в приложении (см. примеры предиктивного интерфейса).

Подумайте об охвате и вау эффекте для клиента, если в рекомендации используются данные не характерные для Тинькофф (характерные: транзакции, события в приложениях, поддержка и тд), опишите схему их получения.

#### Решение

1. Если человек находится в каком-то определенном ТЦ, то ему приходит из приложения пуш со списком магазинов в данном ТЦ (и их расположением, например, этаж) в которых у клиента имеется повышенный кэшбэк. По геопозиции мобильного устройства клиента с установленным мобильным приложением Тинькофф.  


2. Если человек находится рядом с банкоматом Тинькофф, то ему приходит из приложения пуш, нажав на который человек в приложении заранее может сформировать QR-код на внесение или получение нужной суммы наличных. Таким образом, дождавшись своей очереди человеку останется только поднести QR-код к банкомату и получить/внести запрошенную сумму. По геопозиции мобильного устройства клиента с установленным мобильным приложением Тинькофф или прямом обмене данных между банкоматом и мобильным устройством клиентов (если банкомат может сканировать и пинговать своей антенной устройства вокруг).  


3. Если человек на главной странице приложения пролистал все доступные stories (или просто листает stories дольше какого-то порога времени, т.е. возможно ему нечего делать или хочется отвлечься), то предлагать клиенту либо бесконечные stories, либо доступ к списку всех имеющихся stories, чтобы клиент продолжал оставаться в приложении / сформировал привычку чаще заходить в него.  


4. Периодически рекомендовать клиенту проводить диверсификацию денежных средств на его счетах, тем самым проявляя "заботу" о клиенте и зарабатывая на увеличении комиссий за обмен валют.
